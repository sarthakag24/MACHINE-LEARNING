```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_train = np.array([0 , 400 , 100 , 50 , 75  , 300 , 210 , 120 , 150 , 250 , 240  , 401 ]) ## i am saying this size of house 
y_train = np.array([0 , 4200 , 1200 , 500 , 600 ,3000 , 2400 , 1000 , 1500 , 3800 , 1000 , 4235 ]) ## this is cost of house 

COST = [] 
ITERATION = []

```

```python
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
plt.scatter(x_train,y_train,marker="x", c = "green" )
plt.title("Housing Prices")
plt.xlabel("Size of House")
plt.ylabel("Prices of House")
plt.show()
##showing how data is scattered here
```

```python
def cost_function(w , b , x , y) : 
 size = x.shape[0] 

 cost = 0 
 for i in range (size):
        f = w*x[i]+b 
        cost += ((y[i] - f) ** 2 )

 return cost/(2*size)
    
```

```python
def compute_gradient(w, b, x_train, y_train):
    m = x_train.shape[0]
    dj_dw = 0.0
    dj_db = 0.0

    for i in range(m):
        f_wb = w * x_train[i] + b     
        error = f_wb - y_train[i]

        dj_dw += error * x_train[i]   
        dj_db += error                 
    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db
```

```python
def gradient_descent(w, b, x_train, y_train):
    alpha = 0.00001
    iterations = 1000000

    for i in range(iterations):
        dj_dw, dj_db = compute_gradient(w, b, x_train, y_train)

        w = w- alpha * dj_dw
        b = b- alpha * dj_db

        if i % 50000 == 0:
            print(f"current_cost = {cost_function(w,b,x_train,y_train)}") 
            COST.append(cost_function(w,b,x_train,y_train))
            ITERATION.append(i)
    
    return w, b
```

```python

w = 19
b = 26
## take any random w , b of your choice first 
```

```python
## check how our model is working without linear  regression 

plt.scatter(x_train, y_train, marker = 'x',color='blue', label='Training Data')


x_line = np.linspace(min(x_train), max(x_train), 100)
y_line = w * x_line + b

plt.plot(x_line, y_line, color='red', label='Regression Line')

plt.xlabel("x_train")
plt.ylabel("y_train")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()




```

You can see it is not working as it should be

```python
## now lets run it 

w ,b = gradient_descent(w , b , x_train , y_train )
print("*")

print(w)
print(b)
```

```python
plt.scatter(x_train, y_train, marker = 'x',color='blue', label='Training Data')

# create line using learned w, b
x_line = np.linspace(min(x_train), max(x_train), 100)
y_line = w * x_line**1 + b

plt.plot(x_line, y_line, color='red', label='Regression Line')

plt.xlabel("x_train")
plt.ylabel("y_train")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```
