```python
import numpy as np 
import math 
import matplotlib.pyplot as plt
```

```python
x_train = np.array([[210, 5, 1, 45], [141, 3, 2, 40], [85, 2, 1, 35]])
y_train = np.array([400 , 300 , 150])
print(f"x = {x_train} , x.shape = {x_train.shape}")
```

```python
m = x_train.shape[0]
n = x_train.shape[1]
alpha = 7e-5
iterations = 1000000
print("m =",m)
print("n =",n )
print(f"learning rate = {alpha}")
```

```python
w = np.ones(n)
b = 1
```

```python
def predict ( x, w , b ) :
    return np.dot(x , w ) + b 
```

```python
def compute_cost(w , b, x_train, y_train) :
    cost = 0 
    for i in range (m) : 
        f = np.dot(x_train[i] , w ) + b  
        cost += (y_train[i] - f)**2 
        
    return cost/(2*m)   
```

```python
def compute_gradient (w , b , x_train , y_train ) :
    dj_dw = np.zeros(n)
    dj_db = 0 
    for i in range(m) : 
        err = np.dot(w,x_train[i]) + b - y_train[i] 
        dj_dw = dj_dw + err*x_train[i]
        dj_db += err
    dj_dw /= m 
    dj_db /= m 

    return dj_dw , dj_db 
    
    
```

```python
def gradient_descent (w , b, x_train , y_train , alpha  , iterations ) : 

    for i in range( iterations ) : 
        dj_dw , dj_db = compute_gradient (w , b , x_train , y_train ) 
        w -= alpha * dj_dw 
        b -= alpha * dj_db
        if (i % 5000 == 0 ) : 
            print(f"current cost = {compute_cost(w,b,x_train,y_train)}     w= {w} b = {b}  ")
    return w , b             
            
        
```

```python
w , b = gradient_descent(w,b,x_train , y_train , alpha , iterations)
```
